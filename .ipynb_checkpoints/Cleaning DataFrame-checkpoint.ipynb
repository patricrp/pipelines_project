```python
#Getting number of views from YouTube requests
for index, row in df.iterrows():
    if row['Visualizaciones']['items']:
        df.loc[index, 'Views'] = row['Visualizaciones']['items'][0]['statistics']['viewCount']
    else:
        df.loc[index, 'Views'] = 0
```

```python
#WebScrapping countries in Spanish
url_sp = 'https://es.wikipedia.org/wiki/Anexo:Pa%C3%ADses_y_territorios_dependientes_por_poblaci%C3%B3n'
data_sp = requests.get(url_sp).text
soup_sp = BeautifulSoup(data_sp, 'html.parser')


def procesaFila(tag):
    m = tag.find_all("td")
    return m[1].find("a").text


table_sp = soup_sp.find_all('table')[0]
table_sp = [procesaFila(fila) for fila in table_sp.find_all("tr")[1:245]]
```

```python
#Web Scrapping countries in English
url_en = 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population'
data_en = requests.get(url_en).text

soup_en = BeautifulSoup(data_en, 'html.parser')


def procesaFile(tag):
    m = tag.find("a")
    return m.text

table_en = soup_en.find_all('table')[0]
table_en = [procesaFile(fila) for fila in table_en.find_all("tr")[1:244]]
```

```python
countries = table_sp+table_en
```
